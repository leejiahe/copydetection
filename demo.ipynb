```python
import os
import csv
import random
from PIL import Image
```

```python
dev_path = os.path.join(os.getcwd(), 'data', 'dev_validation_set.csv')
dev_csv = []

with open(dev_path , 'r') as csvfile:
    csvreader = csv.reader(csvfile, delimiter = ',')
    for row in csvreader:
        if int(row[2]) == 1:
            dev_csv.append((row[0], row[1]))
```

```python
random_samples = random.sample(dev_csv, k = 6)
```

```python
random_samples
```

```python
references_dir = os.path.join(os.getcwd(), 'data', 'references')
dev_queries_dir = os.path.join(os.getcwd(), 'data', 'dev_queries')
```

```python
idx = 5
references_image_name = random_samples[idx][0]
print(references_image_name)

references_path = os.path.join(references_dir, references_image_name)

Image.open(references_path)
```

```python
idx = 5
dev_queries_image_name = random_samples[idx][1]

dev_queries_path = os.path.join(dev_queries_dir, dev_queries_image_name)

Image.open(dev_queries_path)
```
